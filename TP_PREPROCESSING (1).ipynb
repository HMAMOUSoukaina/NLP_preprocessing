```python
#mise en miniscule
import pandas as pd

# Créer une copie de `data` pour le nouveau DataFrame
data2 = data[['tweet_id', 'text']].copy()

# Convertir le texte en minuscules et le placer dans la colonne `text_cleaned`
data2['text_cleaned'] = data2['text'].str.lower()
# Modifier les options d'affichage pour afficher toutes les colonnes sur une seule ligne
pd.set_option('display.max_columns', None)

# Afficher le nouveau DataFrame avec les trois colonnes tweet_id, text, text_cleaned
print(data2.to_string( index=False))

```

```python
# Suppression des ponctuations  des émojis , émoticônes, url
import pandas as pd
import re



# Créer une copie de data2
data2 = data[['tweet_id', 'text']].copy()

# Supprimer la ponctuation de la colonne 'text_cleaned'
data2['text_cleaned'] = data2['text'].str.lower().apply(lambda x: re.sub(r'[^\w\s]|http\S+|www.\S+|[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F780-\U0001F7FF\U0001F800-\U0001F8FF\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U00002639]', '', x))
# affichage
print(data2.to_string( index=False))


```

```python
# suppression des stopwords
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Télécharger la liste des stopwords pour l'anglais
nltk.download('stopwords')

# Créer une copie de data2
data2 = data2[['tweet_id', 'text', 'text_cleaned']].copy()

# Supprimer les mots vides de la colonne 'text_cleaned' tout en conservant les URLs, les liens et les emojis
stop_words = set(stopwords.words('english'))
data2['text_cleaned'] = data2['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words and not word.startswith('http') and not word.startswith('@') and not any(char.isdigit() for char in word)]))

# Afficher le dataframe
print(data2.to_string(index=False))


```

```python
# Suppression des mots fréquents : Identifiez et supprimez les mots extrêmement fréquents qui pourraient être peu informatifs.
from collections import Counter

# calculer la fréquance des mots dans la colonne  'text_cleaned'
words_frequence = Counter(' '.join(data2['text_cleaned']).split())

#  pour récupérer les 15 mots les plus fréquents
words = words_frequence .most_common(15)

# regrouper ces mots fréquents dans une liste pour les éliminer de notre colonne 'text_cleaned'
words_list = [word for word, _ in words]

# suppression des mots fréquents
data2['text_cleaned'] = data2['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in words_list]))

# affichage du dataframe
print(data2.to_string( index=False))
```

```python
#Suppression des mots rares : Éliminez les mots qui apparaissent très rarement

from collections import Counter

# compter la fréquence des mots
word_frequence = Counter(' '.join(data2['text_cleaned']).split())

# identifer les mots rares qui n'apparaissent qu'une seule fois
mots_rares = [word for word, freq in word_frequence.items() if freq == 1]

# suppression des mots rares de notre colonne 'text_cleaned'
data2['text_cleaned'] = data2['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in mots_rares]))

# affichage du dataframe
print(data2.to_string( index=False))

```

```python
# Stemming : Réduisez les mots à leur racine pour simplifier le texte.
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk
nltk.download('punkt')



# initialisation du stemming
stemmer = PorterStemmer()

# appliquation du sur la colonne 'text_cleaned'
data2['text_cleaned'] = data2['text_cleaned'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))

# affichage du dataframe
print(data2.to_string( index=False))

```

```python
# Lemmatisation : Convertissez les mots à leur forme de base en utilisant le contexte lexical.

import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# NLTK
nltk.download('punkt')
nltk.download('wordnet')

#  le lemmatizer WordNet
lemmatizer = WordNetLemmatizer()

#  appliquer lemmatizer
data2['text_cleaned'] = data2['text_cleaned'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(x)]))

# affichage du dataframe
print(data2.to_string( index=False))
```
